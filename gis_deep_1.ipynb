```python
import csv
import numpy as np
with open('/content/drive/MyDrive/abalone_mini.csv') as csvfile:
  csvreader = csv.reader(csvfile)
  rows = []
  for row in csvreader:
    rows.append(row)

data = np.zeros([5,11])
for n,row in enumerate(rows):
  if row[0] == 'M' : data[n,0] = 1
  if row[0] == 'F' : data[n,0] = 1
  if row[0] == 'I' : data[n,0] = 1
  data[n,3:] = row[1:] #data에 값이 들어있음

def main_exec(x,y):
  model_init()
  run_train(x,y)

input_cnt  = 10   #독립변수
output_cnt = 1   #종속변수
RND_MEAN   = 0
RND_STD    = 1

def model_init():   #weight, bias값들을 정의하는 함수
  global weight, bias
  weight = np.random.normal(RND_MEAN,RND_STD, size = [input_cnt,output_cnt])
  bias = np.random.normal(RND_MEAN, RND_STD, size = [output_cnt])
  
def forward_neuralnet(x):   #y_hat(예상값 구하기)
  y_hat = np.matmul(x,weight) + bias
  return y_hat

def forward_postproc(output, y):    #y^값(예측 값) - y값(실제 값)의 제곱의 평균 값
  diff = output - y
  square = np.square(diff)
  mse = np.mean(square)
  print("mse : \n", mse)
  return mse

def run_train(x,y):
  output = forward_neuralnet(x)   #output에는 y_hat값이 들어감
  loss = forward_postproc(output, y) #loss에는 mse값이 들어감
  return output, loss

main_exec(data[:,:-1], data[:, -1:])

# mse =  To calculate the MSE, 
#        you take the difference between your model's predictions and 
#        the ground truth, square it, and average it out across the whole dataset.
```
